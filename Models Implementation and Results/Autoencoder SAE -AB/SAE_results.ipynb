```python
# Find the paramters (bottleneck size and image size) which lead to the best result for every classifier
measures=['accuracy','f1','precision','recall']
classifs=['adaboost','catboost','lr','rfc','svc','xgb']

results = np.zeros((len(classifs),len(measures),4,2))
i_m=0
b_values=range(40,130,20)
s_values=range(5,16,2)
for measure in measures:
  print(measure)
  i_c=0
  for classif in classifs:
    print(classif)
    train_wo_data =  np.loadtxt('/content/drive/My Drive/Weak learners/programming/Artur/results/wo_coordinates/'+ measure+'_train_'+classif+'.csv',dtype=float,delimiter=',')
    test_wo_data = np.loadtxt('/content/drive/My Drive/Weak learners/programming/Artur/results/wo_coordinates/'+ measure+'_test_'+classif+'.csv',dtype=float,delimiter=',')
    train_w_data =  np.loadtxt('/content/drive/My Drive/Weak learners/programming/Artur/results/w_coordinates/'+ measure+'_train_'+classif+'.csv',dtype=float,delimiter=',')
    test_w_data = np.loadtxt('/content/drive/My Drive/Weak learners/programming/Artur/results/w_coordinates/'+ measure+'_test_'+classif+'.csv',dtype=float,delimiter=',')
    summ=np.zeros((len(b_values),len(s_values),4))

    for b in range(0,len(b_values)):
      for s in range(0,len(s_values)):
        summ[b,s,0]=train_wo_data[b,s]
        summ[b,s,1]=test_wo_data[b,s]
        summ[b,s,2]=train_w_data[b,s]
        summ[b,s,3]=test_w_data[b,s]
        
    i_j = np.where(summ[:,:,0] == np.amax(summ[:,:,0]))
    results[i_c,i_m,0,0]=b_values[i_j[0][0]]
    results[i_c,i_m,0,1]=s_values[i_j[1][0]]
    print(str(measure)+'_wo_train max for : b = '+str(b_values[i_j[0][0]])+', s = '+str(s_values[i_j[1][0]]))
    i_j = np.where(summ[:,:,1] == np.amax(summ[:,:,1]))
    results[i_c,i_m,1,0]=b_values[i_j[0][0]]
    results[i_c,i_m,1,1]=s_values[i_j[1][0]]
    print(str(measure)+'_wo_test max for : b = '+str(b_values[i_j[0][0]])+', s = '+str(s_values[i_j[1][0]]))
    i_j = np.where(summ[:,:,2] == np.amax(summ[:,:,2]))
    results[i_c,i_m,2,0]=b_values[i_j[0][0]]
    results[i_c,i_m,2,1]=s_values[i_j[1][0]]
    print(str(measure)+'_w_train max for : b = '+str(b_values[i_j[0][0]])+', s = '+str(s_values[i_j[1][0]]))
    i_j = np.where(summ[:,:,2] == np.amax(summ[:,:,2]))
    results[i_c,i_m,3,0]=b_values[i_j[0][0]]
    results[i_c,i_m,3,1]=s_values[i_j[1][0]]
    print(str(measure)+'_w_test max for : b = '+str(b_values[i_j[0][0]])+', s = '+str(s_values[i_j[1][0]]))
    print('-------------------------')
    i_c=i_c+1
  i_m=i_m+1
```
