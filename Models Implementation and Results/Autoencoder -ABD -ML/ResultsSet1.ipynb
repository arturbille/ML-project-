```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
res3  = pd.read_csv('results3.csv')
res5  = pd.read_csv('results5.csv')
res7  = pd.read_csv('results7.csv')
res9  = pd.read_csv('results9.csv')
res11 = pd.read_csv('results11.csv')
res13 = pd.read_csv('results13.csv')
res15 = pd.read_csv('results15.csv')
```

```python
res3.drop(columns=['0'],inplace=True)
res5.drop(columns=['0'],inplace=True)
res7.drop(columns=['0'],inplace=True)
res9.drop(columns=['0'],inplace=True)
res11.drop(columns=['0'],inplace=True)
res13.drop(columns=['0'],inplace=True)
res15.drop(columns=['0'],inplace=True)
```

```python
res5.head()
```

```python
titles = ['AdaBoost','XGBoost','CatBoost','RandomForest','SVC','LogisticRegression']
butneck = list(range(40,130,20))
fig, ax = plt.subplots(3,2,figsize=(15,7),sharex=True,sharey=True,dpi=200)
ax = ax.flatten()
plt.tight_layout()
fig.suptitle('Cut_size=3',y=1.1)
for i in range(len(ax)):
  ax[i].plot(butneck, res3.iloc[i*5,1:].astype(float),label= f'{titles[i][:4]}_train_acc')
  ax[i].plot(butneck, res3.iloc[1+(i*5),1:].astype(float),label= f'{titles[i][:4]}_test_acc')
  ax[i].plot(butneck, res3.iloc[2+(i*5),1:].astype(float),label= f'{titles[i][:4]}_train_f1')
  ax[i].plot(butneck, res3.iloc[3+(i*5),1:].astype(float),label= f'{titles[i][:4]}_test_f1')
#   ax[i].fill_between(butneck, res5.iloc[2+(i*5),1:].astype(float),res5.iloc[3+(i*5),1:].astype(float), color='#539ecd', alpha=.1)
  ax[i].set_title(titles[i])
  ax[i].yaxis.grid(linestyle=':')
  ax[i].legend(ncol=4)
plt.show()
```
